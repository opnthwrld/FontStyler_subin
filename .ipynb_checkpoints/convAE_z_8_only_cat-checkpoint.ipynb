```python
#font_size = 128*128, doc2vec_size = 20, letter_vec_size =128

class AE_conv(nn.Module):
    def __init__(self, font_size=128*128, doc2vec_size =20, letter_vec_size =1):
        super(AE_conv, self).__init__()
        #embedded_dim = 16 + doc2vec_size(20) + letter_vec_size(128)
        
        self.Encoder = Encoder_conv_only_cat(img_dim = 1, conv_dim = 64)
        self.Decoder = Decoder_conv_only_cat(img_dim = 1, embedded_dim = 28, conv_dim =64)
        
    
    def forward(self, x_font, doc2vec_vector, batch_size):
        
        origin_shape = x_font.shape
        
        z_latent = self.Encoder(x_font)
        
        z = torch.cat((z_latent.reshape(batch_size, 8), doc2vec_vector),dim=1)
        
        x_hat = self.Decoder(z.reshape(batch_size, 28, 1,1))
        x_hat = x_hat.view(origin_shape)
        
        return x_hat
```

```python
generator = AE_conv().cuda()
```

```python
bce_criterion = nn.MSELoss().cuda()
mse_criterion = nn.BCELoss().cuda()
l1_criterion = nn.L1Loss().cuda()
```

```python
g_optimizer = torch.optim.Adam(generator.parameters(), lr=lr)
```

```python
def train_generator(batch_size, doc2vec_vector, images, generator):
    
    g_optimizer.zero_grad()
   
    #get output from generator and discriminator   
    fake_images = generator(images, doc2vec_vector,batch_size) 
        
    g_loss= F.l1_loss(fake_images, images)
    
    g_loss.backward()
    g_optimizer.step()
    
    return g_loss
```

```python
def train_generator_v(batch_size, doc2vec_vector, images, generator):
    
    #get output from generator and discriminator   
    fake_images = generator(images, doc2vec_vector,batch_size) 
        
    g_loss= F.l1_loss(fake_images, images)
    
    return g_loss
```

```python
np.random.seed(seed=304)
random_table = np.random.normal(size=[2350,128])
```

```python
# set number of epochs and initialize figure counter
num_epochs = 70
num_batches = len(train_loader)
num_fig = 0

#final_g_total_loss = []

#final_g_total_loss_v = []

for epoch in range(num_epochs):
    temp_g_total_loss = []
    
    temp_g_total_loss_v = []

    if epoch % 1 == 0:
        print('----------------------------%d epoch----------------------------------'%(epoch+1))
    
    for n, (vectors, font) in enumerate(train_loader):
        doc2vec_vector = vectors['font_doc2vec']
        letter_vector = vectors['word_index']

        images, letter_vector = font.float().to(device), letter_vector.float().to(device)
        doc2vec_vector = doc2vec_vector.float().to(device)
        temp_batch_size = letter_vector.size()[0]
            
        # Train the generator
        g_loss = train_generator(temp_batch_size, doc2vec_vector, images, generator)
                 
        temp_g_total_loss.append(g_loss.item())
        

        if n % 1000 == 0:
            final_g_total_loss.append(np.mean(temp_g_total_loss))
            
            print('Epoch [%d/%d], Step[%d/%d], g_loss: %.4f' 
                  %(epoch + 1, num_epochs, n+1, num_batches, np.mean(final_g_total_loss)))
        
    # validation part
    for n, (vectors, font) in enumerate(valid_loader):
        
        doc2vec_vector = vectors['font_doc2vec']
        letter_vector = vectors['word_index']

        images, letter_vector = font.float().to(device), letter_vector.float().to(device)
        doc2vec_vector = doc2vec_vector.float().to(device)
        temp_batch_size = letter_vector.size()[0]
        
        # Train the generator
        g_loss = train_generator_v(temp_batch_size, doc2vec_vector, images, generator)

        
        temp_g_total_loss_v.append(g_loss.item())



        if n % 1000 == 0:

            final_g_total_loss_v.append(np.mean(temp_g_total_loss_v))

            print('Epoch [%d/%d], Step[%d/%d], valid_g_loss: %.4f ' 
                  %(epoch + 1, num_epochs, n+1, num_batches, np.mean(final_g_total_loss_v)))

        
```
